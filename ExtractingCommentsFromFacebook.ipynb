```python
from Info1 import business_page_id,page_id,access_token
import requests
import pandas as pd
import json
```

Creating variable comment so that it can work in the f' statement that will pull the
request of URL from Facebook Graph API

```python

comments = '{comments.limit(500)}'
```

URL below helps to pull the maximum amount of comments from the Year 2024 without
errors. Feel free to go to Facebook Graph API to debug

```python
url = f'https://graph.facebook.com/v19.0/{page_id}?fields=posts{comments}&since1706745600&access_token={access_token}'
```

```python
response = requests.request("GET", url)
```

```python
print(response.text.encode('utf-8'))
```

```python
data = json.loads(response.text)
```

```python
ind_comments_df = pd.DataFrame(columns=['comment_id', 'message', 'user_name', 'user_id', 'created_time'])
```

```python
data
```

```python
comments_data = data['posts']
```

```python
for j in range(len(comments_data['data'])):
      if 'comments' in comments_data['data'][j].keys():
           for i in range(len(comments_data['data'][j]['comments']['data'])):
               dict_ = { 'comment_id' : comments_data['data'][j]['comments']['data'][i]['id'],
            'message' : comments_data['data'][j]['comments']['data'][i]['message'],
            'user_name' : comments_data['data'][j]['comments']['data'][i]['from']['name'],
            'user_id': comments_data['data'][j]['comments']['data'][i]['from']['id'],
            'created_time' : comments_data['data'][j]['comments']['data'][i]['created_time'],
            'post_id' : comments_data['data'][j]['id']
            }
               ind_comments_df = ind_comments_df.append(dict_, ignore_index=True)
      else: None
    #print(ind_comments_df)
```
